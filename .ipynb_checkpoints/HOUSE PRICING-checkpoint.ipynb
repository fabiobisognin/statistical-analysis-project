```python
#junto mes y año para crear una columna nueva como fecha de venta

df['Sell Date']=pd.to_datetime(df.YrSold.astype(str) + '/' + df.MoSold.astype(str) + '/01')

df['Sell Date'].min(), df['Sell Date'].max()
```

VENTAS PER MES EN $ - PROGNOSTICO DE VENTAS

```python
#regresión de ventas por mes
matplotlib.rcParams['axes.labelsize'] =14
matplotlib.rcParams['xtick.labelsize'] =12
matplotlib.rcParams['ytick.labelsize'] =12
fig = plt.figure(facecolor='w')
plt.style.use('tableau-colorblind10')
y.plot(figsize=(18,6),marker="o",markersize=5,markerfacecolor='red')
```

```python

matplotlib.rcParams['figure.figsize'] =11,8
decomposition = sm.tsa.seasonal_decompose(y,model='additive')
fig =decomposition.plot()
plt.show()
```

```python
import itertools

p = d = q =range(0,2)

pdq=list(itertools.product(p,d,q))

seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p,d,q))]
print('Ejemplos de combinaciones de parametros para ARIMA')
print(f'SARIMAX:{pdq[1]} x {seasonal_pdq[1]}')
print(f'SARIMAX:{pdq[1]} x {seasonal_pdq[2]}')
print(f'SARIMAX:{pdq[2]} x {seasonal_pdq[3]}')
print(f'SARIMAX:{pdq[2]} x {seasonal_pdq[4]}')
```

```python
from statsmodels.tsa.stattools import adfuller
dftest = adfuller(sales.SalePrice,autolag=None)
dftest
print(f'Prueba estadistica {dftest[0]}')
print(f'P-value  {dftest[1]}')
for k, v in dftest[4].items():
    print('\t{}:{} Los datos son {} estacionarios con {}% de confianza'.format(k, v, "not" if v < dftest[0] else "", 100-int(k[:-1])))

```

```python
import statsmodels.api as sm
#AIC (Akaike Information Criteria)
# Es un estimador de la calidad de los modelos estadisticos. 
#Nos va a decir que parámetros de SARIMA son mejores para nuestros dataset.
#AIC, el cual, entre más pequeño sea, mejor parámetro para ARIMA.

for param in pdq:
    for param_estacionales  in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(y,
                                            order=param,
                                            seasonal_order=param_estacionales,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)
            results = mod.fit()
            print(f'ARIMA {param} x {param_estacionales}12 - AIC:{results.aic}')
        except:
            continue
```

```python
mod = sm.tsa.statespace.SARIMAX(y,
                                            order=(1,1,1),
                                            seasonal_order=(1, 1, 0, 12),
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)
results = mod.fit()
```

```python
#Checo la calidad de los datos antes de la proyección
results.plot_diagnostics(figsize=(16,8))
plt.show()
```

El nivel de error de los datos observados es aceptable (correlograma) para hacer una predicción 

```python
#predicciones observadas
predicciones = results.get_prediction(start=pd.to_datetime('2009-01-01'))
pred_sales = predicciones.conf_int()
pred_sales 
```

```python
#plot prognostico comparativo
ax = y['2006':].plot(label='observados')
predicciones.predicted_mean.plot(ax=ax, label='Pronóstico comparativo', alpha=.7, figsize=(14,7))
ax.fill_between(pred_sales.index,
                pred_sales.iloc[:,0],
                pred_sales.iloc[:,1],
                color='gray',
                alpha=0.2)
ax.set_xlabel('Fecha')
ax.set_ylabel('Ventas de inmuebles')
plt.legend()
plt.show()
```

La predicción funciona bien excepto por los útimos dos meses, donde hay una bajada irregular contra la tendencia estacional de los años anteriores

```python
#Hago un forecast de ventas a 12 meses
prediccion_futuro = results.get_forecast(steps=12)
pred_sales = prediccion_futuro.conf_int()
ax = y.plot(label='Pasado', figsize=(14,7))
prediccion_futuro.predicted_mean.plot(ax=ax, label='Pronostico')
ax.fill_between(pred_sales.index,
                pred_sales.iloc[:,0],
                pred_sales.iloc[:,1],
                color='k',
                alpha=0.2)
ax.set_xlabel('Fecha')
ax.set_ylabel('Ventas de inmuebles')
plt.legend()
plt.show()
```

CONCLUSIÓN: Hay más probabilidades de que el volumen de venta baje aunque no habría que considerar como confiable esta proyección porque baja el valor cero.
